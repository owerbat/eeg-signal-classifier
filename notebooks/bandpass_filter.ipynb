```python
import numpy as np
import pandas as pd
from scipy import signal
from plotly import express as px
```

```python
left_real_info = pd.read_parquet('../data/left_real_info.parquet')
left_real_times = np.load('../data/left_real_time.npy')
left_real_trials = np.load('../data/left_real_trial.npy')
```

```python
time, trial = left_real_times[0], left_real_trials[0]

px.line(x=time, y=trial)
```

```python
sos = signal.butter(N=1, Wn=(1, 50), btype='bandpass', output='sos', fs=len(trial) / 10)
filtered = signal.sosfilt(sos, trial)

px.line(x=time, y=filtered)
```
